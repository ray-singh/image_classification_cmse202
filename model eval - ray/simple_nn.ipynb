```python
import torch
from torch import nn
import torchvision
from torchvision.transforms import Resize
from torchvision.transforms.v2 import ToDtype, Lambda
from torchvision.io import read_image
import numpy as np
import matplotlib.pyplot as plt
import glob

class CancerDataset(torch.utils.data.Dataset):
    def __init__(self, img_labels, img_paths, transform=None, target_transform=None):
        """
        Initialize the CancerDataset object.

        Params::
        - img_labels: Array of labels for the images (0 for non-IDC, 1 for IDC).
        - img_paths: List of file paths to the images.
        - transform: Transformation function for image preprocessing.
        - target_transform: Transformation function for label preprocessing 
        """
        self.img_labels = img_labels
        self.img_paths = img_paths
        self.transform = transform
        self.target_transform = target_transform
        
    def __len__(self):
        # Return the total number of samples in the dataset
        return len(self.img_paths)
    
    def __getitem__(self, idx):
        """
        Get a sample from the dataset at a given index.
        
        Params:
        - idx: Index of the sample to fetch.

        Returns:
        - image: Preprocessed image tensor.
        - label: Transformed label tensor.
        """
        # Get the image path and read the image
        img_path = self.img_paths[idx]
        image = read_image(img_path)
        
        # Fetch the label corresponding to the image
        label = self.img_labels[idx]

        if self.transform:
            image = self.transform(image)
        
        if self.target_transform:
            label = self.target_transform(label)
        
        return image, label

# Define the neural network model
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        # Flatten layer to reshape the input image tensor
        self.flatten = nn.Flatten()

        # Define a sequential stack of linear layers with ReLU activations
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(50*50*3, 1000),  # Input layer (50x50 RGB image flattened to a vector of size 7500)
            nn.ReLU(),                 # ReLU activation function
            nn.Linear(1000, 100),      # Hidden layer with 100 neurons
            nn.ReLU(),
            nn.Linear(100, 2)          # Output layer with 2 neurons (binary classification: non-IDC, IDC)
        )

    def forward(self, x):
        # Flatten the input tensor
        x = self.flatten(x)

        # Pass the input through the sequential layers
        logits = self.linear_relu_stack(x)
        return logits

# Instantiate the neural network model and move it to MPS (Apple Silicon GPU) for faster computation
model = NeuralNetwork().to('mps')

# Load the pre-trained model weights
model.load_state_dict(torch.load('model_weights.pth', weights_only=True))
model.eval()

# Fetch all image file paths recursively from the dataset directory
image_paths = np.array(glob.glob('/Users/calvindejong/Downloads/cancer_images/IDC_regular_ps50_idx5/**/*.png', recursive=True))
labels = np.zeros(len(image_paths), dtype=int)

# Extract labels from file names (class is indicated by the last character before the '.png' extension)
for i in range(len(image_paths)):
    label = image_paths[i][-5]  # The label is stored as the 5th last character in the file name
    labels[i] = int(label)

# Define image transformations (resize to 50x50 and convert to float type tensor)
img_transforms = torchvision.transforms.Compose([
    Resize((50, 50)),           # Resize image to 50x50 pixels
    ToDtype(torch.float, scale=True)  # Convert image to float and scale pixel values (0 to 1)
])

# Create a dataset object
dataset = CancerDataset(
    img_labels=labels,            # Array of image labels
    img_paths=image_paths,        # Array of image file paths
    transform=img_transforms,     # Apply image transformations
    target_transform=Lambda(lambda y: torch.zeros(2, dtype=torch.float).scatter_(
        dim=0, index=torch.tensor(y), value=1))
)

# Create a DataLoader for batching and shuffling data
dataloader = torch.utils.data.DataLoader(dataset, batch_size=20000, shuffle=True)

# Fetch a single batch of data (X: images, y_raw: one-hot encoded labels)
(X, y_raw) = next(iter(dataloader))
```

```python
# Move tensors to MPS
X, y_raw = X.to('mps'), y_raw.to('mps')

m = torch.nn.Softmax(dim=1)
pred_y = m(model(X)).argmax(1)

# Move tensors back to CPU
X, y_raw = X.cpu(), y_raw.cpu()

# Create a figure for displaying images
fig, ax = plt.subplots(4, 4)
plt.title("Images with their corresponding labels")
plt.tight_layout()

# Display 16 images in a 4x4 grid
for i in range(4):
    for j in range(4):
        plt.subplot(4, 4, 4*i+j+1)  # Create a subplot at position (i, j)
        plt.imshow(X[4*i+j].permute(1, 2, 0))  
        plt.title(f"Act: {y_raw[4*i+j].argmax().item()} Pred.: {pred_y[4*i+j].item()}")
        plt.axis('off')  # Turn off axis for better image display

plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import torch

# Move tensors to CPU
y_raw_cpu = y_raw.cpu()
pred_y_cpu = pred_y.cpu()

# Compute confusion matrix
cm = confusion_matrix(y_raw_cpu.argmax(1).numpy(), pred_y_cpu.numpy())

plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import auc

precision, recall, thresholds = precision_recall_curve(y_raw_cpu.argmax(1).numpy(), pred_y_cpu.numpy())
pr_auc = auc(recall, precision)

plt.figure(figsize=(6, 6))
plt.plot(recall, precision, color='b', label=f'PR AUC = {pr_auc:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.show()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, roc_thresholds = roc_curve(y_raw_cpu.argmax(1).numpy(), pred_y_cpu.numpy())

roc_auc = roc_auc_score(y_raw_cpu.argmax(1).numpy(), pred_y_cpu.numpy())
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, color='r', label=f'ROC AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='k')  # Diagonal line (random classifier)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_raw_cpu.argmax(1).numpy(), pred_y_cpu.numpy())
print(f"Accuracy: {accuracy:.2f}")
```
